In this notebook, we will perform exploratory data analysis to understand the underlying patterns, distributions, and relationships between variables in the Global Superstore dataset. This will guide future SQL analysis and dashboard creation in Power BI.

### Univariate Analysis

```python
df['Sales'].hist(bins=30)
plt.title('Distribution of Sales')
plt.show()
```

```python
df['Profit'].hist(bins=30)
plt.title('Distribution of Profit')
plt.show()
```

```python
df['Quantity'].hist(bins=30)
plt.title('Distribution of Quantity')
plt.show()
```

##### Countplot of Categorical Features 

```python
sns.countplot(x='Category', data=df)
plt.title('Category Distribution')
plt.show()
```

```python
sns.countplot(x='Segment', data=df)
plt.title('Segment Distribution')
plt.show()
```

```python
sns.countplot(x='Region', data=df)
plt.title('Region Distribution')
plt.xticks(rotation=90)
plt.show()
```

## Key Insights:

- Sales are right-skewed, indicating a few high-value orders.
- Profit has both positive and negative values, suggesting some losses.
- Most orders have a small quantity (1-3 items).

- Office Supplies category has the highest number of orders.
- The Central region shows the highest number of orders among all regions.
- Consumer segment contributes to most of the purchases.

### Bivariate Analysis

##### Sales vs Profit

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

##### Insight for Sales vs Profit:
While there is generally a positive relationship between sales and profit meaning higher sales often lead to higher profits there are several instances where high sales are associated with negative profits. This suggests that some high-value sales transactions might be heavily discounted or have high associated costs, leading to losses.

#### Region vs Total Sales

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=region_sales, x='Sales', y='Region', palette='viridis')
plt.title('Region vs Total Sales')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.show()
```

##### Insight:
Central Region brings most Revenue overall

#### Category vs Profit

```python
category_profit = df.groupby('Category')['Profit'].sum().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=category_profit, x='Profit', y='Category')
plt.title('Category vs Profit')
plt.xlabel('Profit')
plt.ylabel('Category')
plt.show()
```

##### Insight for Category vs Profit:
Although all categories contribute to overall sales, not all of them are equally profitable. Some categories may have high sales but low or even negative profits, indicating potential issues like high discounting or low margins. Among all categories, **Technology** stands out by generating the highest profits.

#### Discount vs Profit

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

##### Insight for Discount vs Profit:
There is a noticeable negative relationship between discount and profit as the discount increases, profits tend to decrease. Higher discounts, while potentially boosting sales volume, often lead to reduced profitability.

### Correlation Analysis

Correlation heatmap between Numerical Features

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### Insight For Correlation Analysis

- Sales and Profit are moderately positively correlated.
- Discount and Profit show a negative correlation, indicating that higher discounts reduce profitability.
- Quantity and Sales have a weak correlation, suggesting selling higher quantities doesn't always mean higher sales revenue.
- There is a strong positive correlation (0.77) between Sales and Shipping Cost, suggesting that higher sales amounts are associated with higher shipping costs.
